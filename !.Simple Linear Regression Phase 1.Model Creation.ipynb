```python
#pandas is a library and contains the function related to the data base and pandas reading as pd
import pandas as pd
```

```python
#dataset is called by read under score csv function under pandas and its assigned to data set as a variable
dataset=pd.read_csv("salary_data.csv")
```

```python
#independent consider as a input variables we need to split separtae ip 
independent=dataset[["YearsExperience"]]
independent
```

```python
#dependent output variable we need to split out separate variable
dependent=dataset[["Salary"]]
dependent
```

```python
#in this train and test set have to split out.so we have to take from anothe library that is sklearn and model selection
from sklearn.model_selection import train_test_split
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#if we run we get result like above this so we need to separate eveting like x=train,test set,y=train,test set,size means how much we want to take,sate consider like origin 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#in moder creation LinearRegression we have to call like regressor then LinearRegression c+p now linear creation consider as a class,so class variables goint to assign in regressor once assigned into regressor .fit() function we need to callthrow access operator we need to pass the materials (x_train,y_train) in this weigh and bias will get calculate
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train) 
```

```python
#coef_its a function
weight=regressor.coef_
weight
```

```python
#intercept_ function variables its saved
bais=regressor.intercept_
bais
```

```python
# eveluation matrix,y_pred variable we have to create, so regressor.predict is a call function ,evaluation matrix should happen in test data so (x_test)only should give
y_pred=regressor.predict(x_test)
```

```python
#r_is a variable
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
